```python
#j'ai un tableau de dictionnaries match
datas = []
colonnes = [];
for c in range(0,sheet.ncols,1):
    colonnes.append(sheet.cell_value(0, c));
for l in range(1,sheet.nrows,1):
    match = {colonnes[0]:sheet.cell_value(l, 0)};
    for c in range(0,sheet.ncols,1):
        match[colonnes[c]] = sheet.cell_value(l, c);
    datas.append(match);    
    
print(datas)
```

```python
#je veux un dictionnary de division(contenant tout les matchs d'une division)

array_of_leagues = {};
league_actuel = '';
for i in range(0,len(datas),1):
    if(datas[i]['Div'] != league_actuel):
        league_actuel = datas[i]['Div']
        array_of_leagues[league_actuel] = [];
    array_of_leagues[league_actuel].append(datas[i])
    
print(array_of_leagues)
```

```python
#je veux un dictionnary de matchs de chaque équipe

matchs_by_teams = {};
temp_teams = {};
for i in range(0,len(datas), 1):
    temp_match = datas[i];
    if(temp_match['HomeTeam'] in temp_teams):
        matchs_by_teams[temp_match['HomeTeam']].append(temp_match);
    else:
        temp_teams[temp_match['HomeTeam']] = '';
        matchs_by_teams[temp_match['HomeTeam']] = [];
        matchs_by_teams[temp_match['HomeTeam']].append(temp_match);
            
    if(temp_match['AwayTeam'] in temp_teams):
        matchs_by_teams[temp_match['AwayTeam']].append(temp_match);
    else:
        temp_teams[temp_match['AwayTeam']] = '';
        matchs_by_teams[temp_match['AwayTeam']] = [];
        matchs_by_teams[temp_match['AwayTeam']].append(temp_match);

        
        


    
```
